# 4th Question Solution :

```python
df_json.columns=df_json.columns.str.lower() # converting title of each column name to lowercase
data_csv.columns=data_csv.columns.str.lower() # converting title of each column name to lower case for typing comfort

# ValueError: You are trying to merge on object and int64 columns. If you wish to proceed you should use pd.concat
# Reason for error : In the first data frame the lg_dist_name is of type : int64 and in the other it is of the type object
# So we need to typecast the object  to int64

df_json["lg_dist_code"] = df_json["lg_dist_code"].astype(int)

column=['state_name','district_name','lg_dist_code','last_updated']

df_4=pd.merge(df_json,data_csv,left_on=column,right_on=column)
print(df_4)
```
