```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
from joypy import joyplot 
```

```python
dataFrame=pd.read_csv('grains_production_india_from_2001_to_2017.csv')
dF1=dataFrame.pivot_table(index='Year')
dF1.head()
```

```python
# len(dataFrame.columns)
# dataFrame.columns
```

```python
# Storing Means of every column in a list
Mean_list=[]
for x in dF1.columns :
    Mean_list.append(dF1[x].mean())
```

```python
# Replacing the Nan values with mean values
i=0
for x in dF1.columns :
    dF1[x]=dF1[x].fillna(Mean_list[i])
    i+=1
```

```python
# First, I would like to plot the point plot using seaborn library
# Point plot : The middle circle is by default mean some might confuse it with median (As it is not clear)
# The lines represent confidence interval some might confuse it with  standard deviation

#Setting the plot's dimension of layout
sns.set(rc={"figure.figsize":(20, 8)}) 
pa1=sns.pointplot(data=dF1,join=False,orient='v',color='green') # Plotting the point plot

# Initially all the x-axis lables were messed up to resolve this
# Rotate them  by 90 degrees their space occupation will be significantly reduced and will be eye pleasing
pa1.set_xticklabels(pa1.get_xticklabels(), rotation=90) 
pa1.set_xlabel("Local Settings/g-Scale") 
pa1.set_ylabel("Commodities")
print()
# dF1.isna()
```

```python
# Plotting with log scale

#Setting the plot's dimension of layout
sns.set(rc={"figure.figsize":(20, 8)}) 
pa1=sns.pointplot(data=dF1,join=False,orient='v',color='green') # Plotting the point plot for all columns

# Initially all the x-axis lables were messed up to resolve this
# Rotate them  by 90 degrees their space occupation will be significantly reduced and will be eye pleasing
pa1.set_xticklabels(pa1.get_xticklabels(), rotation=90) 
pa1.set_xlabel("Log-Scale") 
pa1.set_yscale('log') # Setting the scale of y-axis as log for better plot
pa1.set_ylabel("Commodities")
print()
# dF1.isna()
```

```python
# Alternative : Instead of tilting just interchange the axes
sns.set(rc={"figure.figsize":(10, 15)})
pa1=sns.pointplot(data=dF1,join=False,orient='h',color='purple')
```

```python
# Box plot is very useful in finding the outliers
# The two black sticks extending from the boxes are called whiskers
# The middle line in box is Median
# THe lower and Upper boundary of box are 1st and 3rd Quartile
# THe horizontal dashes at the extremes are Maximum and Minimum value

# Plotting the plot for the  columns [10,14]
sns.set(rc={"figure.figsize":(20, 8)}) 
pa1=sns.pointplot(data=dF1.iloc[:,10:15],join=False,orient='v',color='green') # Plotting the point plot

# Initially all the x-axis lables were messed up to resolve this
# Rotate them  by 90 degrees their space occupation will be significantly reduced and will be eye pleasing
pa1.set_xticklabels(pa1.get_xticklabels(), rotation=90) 
# pa1.set_xlabel("Log-Scale") 
pa1.set_ylabel("Commodities")
print()
```
