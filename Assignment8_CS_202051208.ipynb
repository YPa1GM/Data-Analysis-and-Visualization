## We can see that Central Limit Theorem holds through by observing the relationship between sample size and mean in the below graphs

```python
arr=np.array(dataFrame.Height)
num = [10, 345, 2400, 5600,10800,15600]
avg = []
for j in num:
    x = [np.mean(np.random.choice(arr, size=j)) for i in range(1000)]
    avg.append(x)
k = 0
fig, ax = plt.subplots(2, 3, figsize =(10, 10))
for i in range(0, 2):
    for j in range(0, 3):
        ax[i, j].hist(avg[k], 10, density = True)
        ax[i, j].set_title(label = "Sample Size : " + str(num[k]))
        k = k + 1
plt.show()
```

```python
epocs=3500
n=5600
x_bar=[]
for i in range(epocs):
    x_bar.append(np.mean(np.random.choice(arr,n,replace=True)))
```

```python
graph=plt.hist(x_bar,bins=250)
no_sample_mean=len(x_bar)
sigma_n=np.std(x_bar)/math.sqrt(no_sample_mean)
print("Computed Standar Error :",sigma_n)
```

## Problem : 3

```python
confidence_Interval=0.95
x_bar=np.sort(x_bar)
Indx=math.floor(no_sample_mean*((1-confidence_Interval)/2))
print("Lower Class Limit : ", x_bar[Indx])
print("Upper Class Limit : ", x_bar[no_sample_mean-Indx-1])
```
