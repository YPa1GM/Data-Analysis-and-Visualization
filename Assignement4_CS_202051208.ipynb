# 1st Question Solution main steps :

```python
merge_DF.columns=pd.MultiIndex.from_tuples([('','States/Uts'),('Egg','2007-08'),('Egg', '2008-09'),
                                            ('Egg', '2009-10'),('Egg', '2010-11'),('Egg', '2011-12'),
                                            ('Milk','2007-08'),('Milk', '2008-09'),('Milk', '2009-10'),
                                            ('Milk', '2010-11'),('Milk', '2011-12')])
merge_DF.columns
```

```python
merge_DF.shape
merge_DF
```

# 2nd Question:

```python
milk_color=sns.color_palette('Set3')
milk_producers=[6, 12, 0, 25, 20] # Storing the indices of the required states
explosion=[0.04,0.01,0.05,0.1,0.06]
presentation_stats=[]
for i in range(5) :
    presentation_stats.append(merge_DF.loc[milk_producers[i],'Milk'].loc['2007-08'])
    
states=[]
for i in range(5) :
    states.append(merge_DF.loc[milk_producers[i],''].loc['States/Uts'])
# print()
plt.pie(presentation_stats,labels=states,colors=milk_color,autopct='%.0f%%',shadow=True,rotatelabels=True,explode = explosion,
       counterclock=True,radius=1.3,startangle = 0,textprops = {'color': 'Purple','size':12},wedgeprops = {'linewidth': 3},
        center = (0.0,0.0))
plt.title('Production of milk (2007-2008)',loc='right',pad=100)
plt.show()
```

# 3rd Question

```python
egg_color=sns.color_palette('rainbow',5)
pie1,pie2,pie3,pie4,pie5=[],[],[],[],[]
Egg_df.columns=milk_df.columns
Egg_df1=Egg_df.set_index('States/Uts',drop=True)#inplace=True)
# Egg_df1.index
# Egg_df1.loc['Gujarat']
for s in states :
    pie1.append(Egg_df1.loc[s,'2007-08'])
    pie2.append(Egg_df1.loc[s,'2008-09'])
    pie3.append(Egg_df1.loc[s,'2009-10'])
    pie4.append(Egg_df1.loc[s,'2010-11'])
    pie5.append(Egg_df1.loc[s,'2011-12'])
    
fig, axes = plt.subplots(3,2,figsize=(20,20))

axes[0,0].set_title("2007-08")
axes[0,1].set_title("2008-09")
axes[1,0].set_title("2009-10")
axes[1,1].set_title("2010-11")
axes[2,0].set_title("2011-12")

axes[0,0].pie(pie1,colors=egg_color,autopct='%.0f%%',shadow=True,textprops={'fontsize': 20},explode=explosion)
axes[0,1].pie(pie2,colors=egg_color,autopct='%.0f%%',shadow=True,textprops={'fontsize': 20},explode=explosion)
axes[1,0].pie(pie3,colors=egg_color,autopct='%.0f%%',shadow=True,textprops={'fontsize': 20},explode=explosion)
axes[1,1].pie(pie4,colors=egg_color,autopct='%.0f%%',shadow=True,textprops={'fontsize': 20},explode=explosion)
axes[2,0].pie(pie5,colors=egg_color,autopct='%.0f%%',shadow=True,textprops={'fontsize': 20},explode=explosion,)
axes[2,1].axis('off')

plt.show()
```
